# Pickle

```python
import pandas as pd
```

```python
data_folder_in = "datos/{}"
data_folder_out = "datos/escritura/{}"

iconos = 'iconos.csv'
pkl = 'iconos.pkl'
```

```python
df = pd.read_csv(data_folder_in.format(iconos), sep=";")
df
```

```python
df.to_pickle(data_folder_out.format(pkl))
```

```python
df = None
```

```python
import pickle
df = pickle.load(open(data_folder_out.format(pkl), "rb" ) )
df
```

```python
pickle.dump(df, open( data_folder_out.format(pkl), "wb" ) )
df = None
```

```python
df = pickle.load(open(data_folder_out.format(pkl), "rb" ) )
df
```
