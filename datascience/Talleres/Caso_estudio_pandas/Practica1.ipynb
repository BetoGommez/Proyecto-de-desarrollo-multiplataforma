```python
df_selected.reset_index(inplace=True)
```

```python
df_selected.head()
```

```python
df = df_selected.copy()
```

```python
df.head()
```

```python
df['Date']
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df.groupby('Month').sum()
```

```python
df['Total']=df['Berri1']+df['Boyer']+df['Boyer 2']+df['Parc']+df['PierDup']
```

```python
df.head()
```

```python
df[['Month','Total']].groupby('Month').sum().plot(kind='bar')
```
