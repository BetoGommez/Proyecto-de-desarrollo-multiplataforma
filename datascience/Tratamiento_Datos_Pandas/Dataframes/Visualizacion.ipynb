# Visualizaci√≥n

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import warnings

from pathlib import Path

data_path= Path("./data")

data = pd.read_csv(data_path / 'house.csv')
warnings.filterwarnings('ignore')
data.head()
```

```python
df = data.drop_duplicates(subset=['date'], keep='first')
df.shape
```

```python
df.index=df['date']
df=df.drop(labels = ['date'], axis=1)
df.head()
```

### Gr√°fico de lineas

```python
df[['price']].plot()
```

```python
df[['bathrooms']].plot(kind='bar' , figsize = [20,8])
```

```python
df[['bathrooms','floors']].plot(kind='bar' , figsize = [20,8])
```

```python
(df['bathrooms']/df['floors']).mean()
```

```python
df.plot(kind = 'scatter', x = 'yr_built', y = 'sqft_above', figsize = (10,8))
```

```python
df[['bathrooms','floors']].plot(kind= 'box',figsize= (20,10))
```
