# Agrupaciones

```python
data.groupby('condition').agg({
    'price' : 'mean',
    'bathrooms' : 'median',
    'waterfront' : 'max',
    'floors' : 'min'
}) 
```

```python
def media_mayores_1_millon(x):
    return np.mean(x[x>1_000_000])
```

```python
data[['price','condition']].groupby('condition')['price'].apply(media_mayores_1_millon)
```
