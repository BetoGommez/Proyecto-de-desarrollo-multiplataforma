# Gráficos para distribuciones

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('titanic')
```

```python
fig, ax= plt.subplots(figsize= (10,5))
sns.displot(data=data['age'],ax=ax,hue_norm=True)
ax.set_title("Distribución edad")
```

```python

sns.displot(data,x="age",hue="sex",height=7,element='step')

```

```python
sns.displot(data, x="age", hue= 'sex', height=7,multiple='stack')
```

```python
data['age']
```

```python
sns.displot(data, x="age", col= 'sex', height=7,stat='density')
```

# KDE

```python
fig, ax=plt.subplots(figsize=(10,5))
sns.displot(data=data['age'], kind="kde", ax=ax)
```

# ECDF

```python
sns.displot(data, x="age", kind="ecdf",hue="sex")
```
