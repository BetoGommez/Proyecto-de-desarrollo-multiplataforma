# Categóricos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('titanic')
data
```

```python
sns.catplot(x='pclass',y='age',data=data, kind='swarm')
```

# Boxplots

```python
fig, ax = plt.subplots(figsize = (10,5))
sns.boxplot(x='pclass',y='age',data=data,ax=ax)
```

```python
sns.catplot(x = 'pclass', y='survived', data = data, kind = 'bar')
```
