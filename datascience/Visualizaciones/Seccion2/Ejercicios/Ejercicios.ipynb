```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')

from pathlib import Path
data_path = Path('../data')

%matplotlib inline

data = pd.read_csv('Chicago_Crimes_2017.csv')
```

```python
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize = (20,8))
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
data['month'] = data['Date'].dt.month

battery_crimes = data[data['PrimaryType'] == "BATTERY"].groupby("month").size();



crimes_by_month = data.groupby('month').size()



ax.bar(x=crimes_by_month.index,height=crimes_by_month.values)
ax1=ax.twinx();
ax1.scatter(x=battery_crimes.index,y=battery_crimes.values, color="red")

ax.set_xticks(crimes_by_month.index.values)
ax.set_title("Nº de crimenes por mes")
ax.set_ylabel("Meses del año")

bar = mpatches.Patch(color="blue", label= "Nº crimenes")
battery = mpatches.Patch(color="blue", label= "Nº crimenes")

plt.legend(handles = [bar, battery], title = "Leyenda")

```
