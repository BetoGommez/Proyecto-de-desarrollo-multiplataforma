```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')

from pathlib import Path
data_path = Path('../data')

%matplotlib inline

data = pd.read_csv(data_path/ 'house.csv')
```

## Gráfico de tartas

```python
cities= data.query('city in ("Seattle", "Renton", "Bellevue", "Redmond", "Issaquah")')
vc = cities['city'].value_counts(normalize=True)
vc
```

```python
fig, ax = plt.subplots(figsize= (15,7))
ax.pie(vc, labels=vc.index, autopct='%1.1f%%')
```

## Gráfico de tablas

```python
df = data.query('condition in (2,3,4) and bedrooms in (1,2,3)')
_ = df.pivot_table(index = 'condition', columns = 'bedrooms', values = 'price', aggfunc=np.count_nonzero)
_
```

```python
price_mean = df.groupby('bedrooms')['price'].mean()
price_mean
```

```python
fig, ax = plt.subplots(figsize= (10,7))
ax.bar(x= price_mean.index, height=price_mean.values)
ax.table(_.values, loc = 'bottom', rowLabels=_.index.values,colLabels=_.columns.values)
plt.subplots_adjust(left=0, bottom=-0.2)
plt.xticks([])
plt.ylabel('Precio medio por nº baños')
```

## Mapa de calor

```python
data.corr()
```

```python
fig, ax = plt.subplots(figsize= (8,6))
ax.pcolormesh(np.arange(13), np.arange(13), data.corr().values)
plt.xticks(np.arange(18),data.columns, rotation = 'vertical');
plt.yticks(np.arange(18),data.columns, rotation = 'horizontal');
```

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')

X=np.arange(-2,2,0.25)
Y=np.arange(-2,2,0.25)
X,Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.cos(R)

surf = ax.plot_surface(X,Y,Z,cmap='Greens')
```
